```python
# Fraud Detection in Internship Applications

import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Create Sample Dataset
# -----------------------------
np.random.seed(42)

n = 100

applications = pd.DataFrame({
    "Application_ID": range(1, n + 1),
    "Submission_Time": np.random.randint(1, 300, n),
    "Duplicate_Count": np.random.randint(0, 3, n),
    "Age": np.random.randint(18, 35, n),
    "CGPA": np.round(np.random.uniform(2.0, 4.0, n), 2)
})

# Add suspicious records
applications.loc[95:, "Submission_Time"] = [1,2,3,4,5]
applications.loc[95:, "Duplicate_Count"] = [5,6,4,7,8]

print("Dataset Preview:")
print(applications.head())

# Step 2: Train Isolation Forest

features = applications[["Submission_Time","Duplicate_Count","Age","CGPA"]]

model = IsolationForest(
    contamination=0.05,
    random_state=42
)

applications["Fraud"] = model.fit_predict(features)

applications["Fraud"] = applications["Fraud"].map({1:"Normal",-1:"Fraud"})

print("\nFraud Detection Results:")
print(applications["Fraud"].value_counts())


# Step 3: Display Fraud Cases

fraud_cases = applications[applications["Fraud"]=="Fraud"]

print("\nSuspicious Applications:")
print(fraud_cases)

# Save results
applications.to_csv("Fraud_Detection_Results.csv", index=False)

# Step 4: Visualization

colors = applications["Fraud"].map({"Normal":"blue","Fraud":"red"})

plt.figure(figsize=(8,5))
plt.scatter(applications["Submission_Time"],
            applications["Duplicate_Count"],
            c=colors)

plt.xlabel("Submission Time")
plt.ylabel("Duplicate Count")
plt.title("Fraud Detection in Internship Applications")
plt.show()

print("\nFraud Detection Completed Successfully!")
```
